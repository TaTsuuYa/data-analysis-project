# Cleaning data

## Import Libraries

```python
import pandas as pd
```

## Import data

```python
df = pd.read_csv('../data/shows.bak.csv')
```

Display data types.

```python
df.info()
```

Display summary statistics.

## Process missing values
Count null values

```python
df.isnull().sum()
```

Clean premiered

```python
df.loc[df['premiered'].isnull(), 'premiered'] = 'Unknown'
df[df['premiered'].isnull()]['premiered'].count()
```

Clean broadcast

```python
df.loc[df['broadcast'].isnull(), 'broadcast'] = 'Unknown'
df[df['broadcast'].isnull()]['broadcast'].count()
```

Clean genres

```python
df.loc[df['genres'].isnull(), 'genres'] = 'Unknown'
df[df['genres'].isnull()]['genres'].count()
```

Clean demographic

```python
df.loc[df['demographic'].isnull(), 'demographic'] = 'Unknown'
df[df['demographic'].isnull()]['demographic'].count()
```

Dlete entries that doesn't have a rating

```python
df = df[df['rating'].notnull()]
```

## Remove puplicates

```python
df = df.drop_duplicates(subset=['title'])
df[df['title'].duplicated()]
```

# Save the Cleaned Dataset
Export the cleaned DataFrame to a new CSV file.

```python
df.to_csv('../data/shows_cleaned.csv', index=False)
```
